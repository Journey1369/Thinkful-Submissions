```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
var1 = np.random.normal(5, .5, 100)
var2 = np.random.normal(10, 1, 100)
var3 = var1+var2
```

```python
bernoulli= np.random.binomial(1, .5, 100)
plt.hist(bernoulli)
plt.show()
```

```python
power = np.random.power(20, 100)
plt.hist(power)
plt.show()
```

```python
triangular = np.random.triangular(1, 50, 100)
plt.hist(triangular)
plt.show()
```

```python
rayleigh = np.random.rayleigh(1, 100)
plt.hist(rayleigh)
plt.show()
```

```python
uniform = np.random.uniform(1, 50, 100)
plt.hist(uniform)
plt.show()
```

```python
pareto = np.random.pareto(1, 100)
plt.hist(pareto)
plt.show()
```

```python
mean = np.mean(var3)
sd = np.std(var3)

plt.hist(var3)
plt.axvline(x=mean,color='black')
plt.axvline(x=mean+sd, color='red')
plt.axvline(x=mean-sd, color='red')
plt.show()
```

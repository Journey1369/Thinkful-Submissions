```python
import pandas as pd
import numpy as np
import os
import matplotlib.pyplot as plt

print(os.getcwd())
```

```python
fname = "athlete_events.csv"
df = pd.read_csv(fname)
```

```python
df.head()
```

```python
%matplotlib inline
plt.plot(df['Age'], color='orange') 
plt.show()
```

__*I believe that in the graph above, I am looking at the variation of age represented by the color Orange. The parameters are set forth by a minimum standard age range 20-100. Furthermore, the opulation density is represented in the age category.*__

```python
plt.plot(df['Height'], color='Fuchsia')
plt.plot(df['Weight'], color='Blue')
plt.show()
```

__*In the graph above, the height & weight data is represented by the colors "Fuchsia" for Height & "Blue" for Weight. The two different categories are seperated by two different plot lines.*__

```python
plt.scatter(y=df['Height'], x=df['Weight'])
plt.ylabel('Height')
plt.xlabel('Weight')
plt.title('Height & Weight')
plt.show()
```

__*The graph above represents Height & Weight within a scatter plot graph. In the graph above, the viewer can see the variations of height & weight with the average being represented in a cluster.*__

```python
plt.figure(figsize=(5, 5))
plt.subplot(1,2,2)
plt.scatter(y=df['Year'], x=df['Season'])
plt.ylabel('Year')
plt.xlabel('Season')
plt.title('Years & Seasons')

plt.tight_layout()
plt.show()
```

__*This graph is a tight layout representing Years & Seasons. The graph depicts when (Years) the data is shown followed by the most common season from the data.*__

```python
plt.figure(figsize=(10, 5))

plt.subplot(1, 2, 1)
plt.plot(df['Height'], color='Fuchsia')
plt.ylabel('Height')
plt.title('Olypmic Height Qualifiers')

plt.subplot(1, 2, 2)
plt.plot(df['Weight'], color='Blue')
plt.ylabel('Weight')
plt.title('Olypmic Weight Qualifiers')
plt.show()
```

__*The above graph depicts two different subplots for two different data sets. Fuchsia represents Olympic Height Qualifiers and Blue depicts Olympic Weight qualifiers and the variations of the data selected.*__

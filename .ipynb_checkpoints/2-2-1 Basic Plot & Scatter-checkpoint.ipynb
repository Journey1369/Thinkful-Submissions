```python
np.random.seed(1221)
df = pd.DataFrame()
df['rand'] = np.random.rand(100)
df['rand_sq'] = df['rand'] ** 2
df['rand_shift'] = df['rand'] + 2
df['counts_sq'] = df.index ** 2 
df['counts_sqrt'] = np.sqrt(df.index)
```

```python
plt.plot(df['rand'])
plt.show()
```

```python
np.random.seed(1221)
df = pd.DataFrame()
df['rand'] = np.random.rand(100)
df['rand_sq'] = df['rand'] ** 2
df['rand_shift'] = df['rand'] + 2
df['counts_sq'] = df.index ** 2 
df['counts_sqrt'] = np.sqrt(df.index)
```

```python
plt.plot(df['counts_sq'])
plt.show()
```

```python
plt.plot(df['counts_sqrt'])
plt.show()
```

```python
plt.plot(df['rand'], color='purple')
plt.ylim([-0.1, 1.1])
plt.ylabel('Values')
plt.title('Random Series')
plt.show()
```

```python
plt.plot(df['rand'], color='fuchsia')
plt.plot(df['rand_shift'], color='blue')
plt.plot(df['rand_sq'], color='green')
plt.ylim([-1.3, 6.9])
plt.ylabel('Home Values')
plt.title('Random Fuchsia Series')
plt.show()
```

```python
plt.plot(df['counts_sq'], color='blue')
plt.ylim([-0.1, 1.3])
plt.ylabel('Covering Mistakes')
plt.title('Mistakes Were Made')
plt.show()
```

```python
plt.scatter(x=df['rand'], y=df['rand_sq'])
plt.show()
```

```python
plt.scatter(
    x=df['rand'],
    y=df['rand_sq'],
    color='purple',
    marker='x', s=10
)
plt.scatter(
    x=df['rand'],
    y=df['rand_shift'],
    color='green',
    marker='x', s=10
)
plt.show()
```

```python
df.plot(kind='scatter', x='counts_sq',y= 'counts_sqrt')
df.plot(kind='line')
plt.show()
```
